```python
import pandas as pd
df = pd.read_csv('titanic.csv')
df
```

```python
df = df.drop(['PassengerId', 'Name', 'SibSp', 'Parch', 'Ticket', 'Cabin', 'Embarked'], axis = 1)
df
```

```python
x = df.Survived
y = df.drop(['Survived'], axis = 1)
x
y
```

```python
x
```

```python
dummies = pd.get_dummies(y.Sex).astype(int)
dummies
```

```python
y = pd.concat([y, dummies], axis = 'columns')
y
```

```python
y = y.drop(['Sex'], axis = 1)
y
```

```python
y = y.drop(['female'], axis = 1)
y
```
